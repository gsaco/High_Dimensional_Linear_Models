## **2 Overfitting (8 points)**

```julia
using Random, Statistics, DataFrames, GLM, StatsBase, Plots
```

### (1) Generación de variables y utilidades

```julia
Random.seed!(3324)
n = 1000
σ = 1.0
X = randn(n)                  
ϵ = σ .* randn(n)
Y = X .+ ϵ             
```

```julia
p_list = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]

function datasplitting(data::DataFrame, proportion)
    sampling = sample(1:size(data,1), Int(proportion*size(data,1)); replace=false)
    sampling = sort(sampling)
    train = data[sampling, :]
    remaining = [i for i in 1:size(data,1) if isempty(searchsorted(sampling, i))]
    test = data[remaining, :]
    return train, test
end
```

### (2) Metricas (MSE y R² y R2 OOS)

```julia
r2_in_sample(y, ŷ) = 1 - mean((y .- ŷ).^2) / mean((y .- mean(y)).^2)
r2_adj(y, ŷ, p) = begin
    n = length(y)
    p < n ? 1 - ( (n/(n - p)) * mean((y .- ŷ).^2) ) / mean((y .- mean(y)).^2) : NaN
end
r2_oos(ytest, ŷtest) = 1 - mean((ytest .- ŷtest).^2) / mean((ytest .- mean(ytest)).^2)
```

```julia
# construir DataFrame con y y potencias de x como en tus celdas
function make_poly_df(x::AbstractVector, y::AbstractVector, p::Int)
    df = DataFrame(y = y)
    for k in 1:p
        df[!, Symbol("x", k)] = x .^ k
    end
    return df
end
```

### (3) Contenedores

```julia
R2_full = Float64[]     # R² in-sample
R2_adj  = Float64[]     # R² ajustado
R2_oos  = Float64[]     # R² out-of-sample
```

### (4) Estimación usando GLM.lm + predict, y split 75/25 con datasplitting

```julia
Random.seed!(2025)  # split fijo
for p in p_list
    df = make_poly_df(X, Y, p)

    # ---- Full sample
    covs = Symbol.("x" .* string.(1:p))
    Xfull = Matrix(select(df, covs))
    m_full = lm(Xfull, df.y)
    ŷ_full = predict(m_full, Xfull)
    push!(R2_full, r2_in_sample(df.y, ŷ_full))
    push!(R2_adj,  r2_adj(df.y, ŷ_full, p))

    # ---- Train/Test 75/25 con tu función
    train, test = datasplitting(df, 0.75)
    Xtr = Matrix(select(train, covs)); ytr = train.y
    Xte = Matrix(select(test,  covs)); yte = test.y
    m_tr = lm(Xtr, ytr)
    ŷ_te = predict(m_tr, Xte)
    push!(R2_oos, r2_oos(yte, ŷ_te))
end
```

### (5) Gráficos

```julia
default(marker=:circle, linewidth=2)
plot(p_list, R2_full; xlabel="Numero de observaciones", ylabel="R²",
     title="R² en muestra completa vs complejidad", xscale=:log10, ylim=(-0.5, 1.0))
```

```julia
plot(p_list, R2_adj;  xlabel="Numero de observaciones", ylabel="R² ajustado",
     title="R² ajustado vs complejidad", xscale=:log10, ylim=(-0.5, 1.0))
```

```julia
plot(p_list, R2_oos;  xlabel="Numero de observaciones", ylabel="R² out-of-sample",
     title="R² out-of-sample (split 75/25) vs complejidad", xscale=:log10, ylim=(-1.0, 1.0))
```

## **3 Real data (9 points)**

```julia
using CSV, DataFrames
df = CSV.read("C:\\Users\\User\\Desktop\\High_Dimensional_Linear_Models\\apartments.csv", DataFrame)
```

# Sección de Limpieza

## Pregunta 1: Creación de la cuadrática del área (0.25 pts)

```julia
df.area2 = df.area .^ 2
```

## Pregunta 2: Conversión en Dummies (0.75 pts)

#### Descripción de DATA

```julia
using DataFrames, Statistics

cols = [:hasparkingspace, :hasbalcony, :haselevator, :hassecurity, :hasstorageroom]

t = describe(df[!, cols], :eltype, :nmissing, :nunique)
t.pct_missing = round.((t.nmissing ./ nrow(df)) .* 100; digits=2)
select!(t, [:variable, :eltype, :nmissing, :pct_missing, :nunique])

t 
```

#### Reemplazo a Dummies

```julia
to01_keepmiss(x) = x === missing ? missing :
                   x isa AbstractString ? (lowercase(x) == "yes" ? 1 : 0) :
                   x isa Bool ? (x ? 1 : 0) :
                   x isa Real ? (x != 0 ? 1 : 0) : 0

for col in [:hasparkingspace, :hasbalcony, :haselevator, :hassecurity, :hasstorageroom]
    df[!, col] = to01_keepmiss.(df[!, col])
end
```

### Pregunta 3: Obtención de la última cifra del área (1 pts)

```julia
lastdigit_area(x) = x === missing ? missing :
    x isa Real ? mod(abs(floor(Int, x)), 10) :
    x isa AbstractString ?
        let v = tryparse(Float64, x)
            v === nothing ? missing : mod(abs(floor(Int, v)), 10)
        end :
    missing

ld = lastdigit_area.(df.area)  
for d in 0:9
    df[!, Symbol("end_$d")] = Int.(coalesce.(ld .== d, false))
end
```

### Pregunta 4: Modelo de Estimación Lineal

```julia
using DataFrames, CategoricalArrays, StatsModels, GLM, Statistics

# 0) Categóricas
for c in [:month, :type, :rooms, :ownership, :buildingmaterial]
    df[!, c] = categorical(df[!, c])
end
```

```julia
# 2) Columnas para el modelo
end_terms = [Symbol("end_$(i)") for i in 0:8]  # omite end_9

dist_cols = [:centredistance, :schooldistance, :clinicdistance, :postofficedistance,
             :kindergartendistance, :restaurantdistance, :collegedistance, :pharmacydistance]

all_cols = vcat([:price, :area, :area2], dist_cols, end_terms,
                [:month, :type, :rooms, :ownership, :buildingmaterial])
```

```julia
# 3) Filtra missings solo en variables usadas
dfreg = dropmissing(df, all_cols)
```

```julia
# 4) Fórmula
f = @formula(price ~ area + area2 +
    centredistance + schooldistance + clinicdistance + postofficedistance +
    kindergartendistance + restaurantdistance + collegedistance + pharmacydistance +
    end_0 + end_1 + end_2 + end_3 + end_4 + end_5 + end_6 + end_7 + end_8 +
    month + type + rooms + ownership + buildingmaterial)
```

```julia
m = lm(f, dfreg)
```

```julia
using Distributions, Printf

function comment_end0(m)
    cn  = coefnames(m)
    idx = findfirst(isequal("end_0"), cn)
    idx === nothing && error("end_0 no está en el modelo.")
    idx = Int(idx)

    β   = coef(m)[idx]
    se  = stderror(m)[idx]
    t   = β / se
    p   = 2 * ccdf(TDist(dof_residual(m)), abs(t))
    sig = p < 0.05 ? "estadísticamente significativo" : "no estadísticamente significativo"

    @printf "Controlando por área, área², distancias y dummies (mes, tipo, rooms, propiedad, material), tener último dígito 0 se asocia con un precio %.2f mayor que el dígito 9. Efecto %s.\n" β sig
end

comment_end0(m)
```

### Pregunta 5: Partición

```julia
# A) OLS completo (base = end_9)
f_full = @formula(price ~ area + area2 +
    centredistance + schooldistance + clinicdistance + postofficedistance +
    kindergartendistance + restaurantdistance + collegedistance + pharmacydistance +
    end_0 + end_1 + end_2 + end_3 + end_4 + end_5 + end_6 + end_7 + end_8 +
    month + type + rooms + ownership + buildingmaterial)

m_full = lm(f_full, dfreg)
```

```julia

# B) FWL: "partialling-out" de todos los controles (sin end_0)
f_y = @formula(price ~ area + area2 +
    centredistance + schooldistance + clinicdistance + postofficedistance +
    kindergartendistance + restaurantdistance + collegedistance + pharmacydistance +
    end_1 + end_2 + end_3 + end_4 + end_5 + end_6 + end_7 + end_8 +
    month + type + rooms + ownership + buildingmaterial)

f_d = @formula(end_0 ~ area + area2 +
    centredistance + schooldistance + clinicdistance + postofficedistance +
    kindergartendistance + restaurantdistance + collegedistance + pharmacydistance +
    end_1 + end_2 + end_3 + end_4 + end_5 + end_6 + end_7 + end_8 +
    month + type + rooms + ownership + buildingmaterial)

r_y = residuals(lm(f_y, dfreg))
r_d = residuals(lm(f_d, dfreg))

m_fwl = lm(@formula(r_y ~ 0 + r_d), DataFrame(r_y=r_y, r_d=r_d))  # sin intercepto
```

```julia
using DataFrames, Printf

tidy_ct(m) = begin
    ct = coeftable(m)
    DataFrame(
        Term     = coefnames(m),
        Coef     = round.(ct.cols[1], digits=3),
        StdError = round.(ct.cols[2], digits=3),
        t        = round.(ct.cols[3], digits=2),
        p        = [@sprintf("%.2e", x) for x in ct.cols[4]],
    )
end

# C) Tablas y verificación
println("=== OLS completo ===");  display(tidy_ct(m_full))
println("=== FWL (parcial) ==="); display(tidy_ct(m_fwl))

idx = findfirst(==("end_0"), coefnames(m_full))
β_full, se_full = coef(m_full)[idx], stderror(m_full)[idx]
β_fwl,  se_fwl  = coef(m_fwl)[1],  stderror(m_fwl)[1]

@printf "\nend_0 OLS: β=%.6f  se=%.6f\n" β_full se_full
@printf "end_0 FWL: β=%.6f  se=%.6f\n" β_fwl  se_fwl
println("coef iguales: ", isapprox(β_full, β_fwl; atol=1e-10))
println("se   iguales: ", isapprox(se_full, se_fwl; atol=1e-10))
```

## Price premium for area that ends in 0-digit

### 1. Entrenamiento del modelo (1.25 pts)

```julia
# Variables del modelo (mismo set pero SIN end_0; base = end_9)
dist_cols = [:centredistance,:schooldistance,:clinicdistance,:postofficedistance,
             :kindergartendistance,:restaurantdistance,:collegedistance,:pharmacydistance]
end_terms = [:end_1,:end_2,:end_3,:end_4,:end_5,:end_6,:end_7,:end_8]
vars = vcat([:price,:area,:area2], dist_cols, end_terms,
            [:month,:type,:rooms,:ownership,:buildingmaterial])
```

```julia
# Train set: apartamentos cuyo área NO termina en 0
train = filter(:end_0 => ==(0), df)
train = dropmissing(train, vars)
```

```julia
# Misma regresión, entrenada en el subconjunto
f_train = @formula(price ~ area + area2 +
    centredistance + schooldistance + clinicdistance + postofficedistance +
    kindergartendistance + restaurantdistance + collegedistance + pharmacydistance +
    end_1 + end_2 + end_3 + end_4 + end_5 + end_6 + end_7 + end_8 +
    month + type + rooms + ownership + buildingmaterial)

m_train = lm(f_train, train)
```

### 2. Predecir Precios (1.25 pts)


```julia
using DataFrames, GLM

vars_pred = vcat([:area,:area2], dist_cols, end_terms,
                 [:month,:type,:rooms,:ownership,:buildingmaterial])

# conserva toda la muestra pero elimina filas con missing en variables del modelo
df_pred = dropmissing(df, vars_pred)

# predicción para toda la muestra (incluye end_0==1)
df_pred.pred_price = predict(m_train, df_pred)
```

```julia
# si quieres comparar con el precio observado
df_pred.err = df_pred.price .- df_pred.pred_price
first(df_pred, 5)
```

### 3. Comparar Promedios (0.50 pts)


```julia
using DataFrames, Statistics, Printf

# Filtra los que terminan en 0 y tienen price/pred_price disponibles
g = df_pred[(df_pred.end_0 .== 1) .&
            .!ismissing.(df_pred.price) .&
            .!ismissing.(df_pred.pred_price), :]

if nrow(g) == 0
    println("No hay filas end_0==1 con predicción disponible.")
else
    μy  = mean(g.price)
    μŷ  = mean(g.pred_price)
    prem = μy - μŷ
    prem_pct = 100 * prem / μŷ

    display(DataFrame(N=nrow(g),
        avg_price=round(μy,digits=2),
        avg_pred =round(μŷ,digits=2),
        premium  =round(prem,digits=2),
        premium_pct=round(prem_pct,digits=2)))

    println(prem > 0 ?
        @sprintf("Hipótesis: se venden por ENCIMA del modelo (≈ %.2f%%).", prem_pct) :
        @sprintf("Hipótesis: NO superan al modelo (≈ %.2f%%).", prem_pct))
end

```
