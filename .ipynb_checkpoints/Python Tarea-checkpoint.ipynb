## **2 Overfitting (8 points)**

### (1) Generación de variables y utilidades

```python
# ============================================================
# (1 punto) Generación de variables y bucle adecuado
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(123)

# Parámetros
n = 1000
sigma = 1.0
p_list = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]

# Proceso generador de datos: intercepto = 0
X = rng.standard_normal(n)                 # variable base
eps = sigma * rng.standard_normal(n)
Y = 1.0 * X + eps                          # Y = X + ruido

# Escalar X a [-1, 1] para estabilizar potencias altas y evitar overflow
X_scaled = X / np.max(np.abs(X))

def make_poly_features(x_unit, p: int):
    """
    Crea matriz de características [x, x^2, ..., x^p] sin intercepto.
    x_unit debe estar en [-1,1] para estabilidad numérica.
    """
    # (n, p) usando potencias acumuladas estables
    # np.power es seguro con |x|<=1; dtype float64
    powers = np.arange(1, p + 1, dtype=np.int32)
    return np.power(x_unit[:, None], powers, dtype=np.float64)

def r2_adjusted(r2: float, nobs: int, k: int):
    """
    R^2 ajustado = 1 - (1-R^2)*(n-1)/(n-k-1)
    Si n - k - 1 <= 0, devuelve np.nan (no definido).
    """
    denom = (nobs - k - 1)
    if denom <= 0:
        return np.nan
    return 1.0 - (1.0 - r2) * (nobs - 1) / denom

# Contenedores de resultados
r2_full_list = []
ar2_full_list = []
r2_oos_list = []
```

```python
# ============================================================
# (1 punto) Estimación en la muestra completa
# ============================================================
for p in p_list:
    Xp = make_poly_features(X_scaled, p)

    # Modelo sin intercepto
    m_full = LinearRegression(fit_intercept=False)
    m_full.fit(Xp, Y)
    yhat_full = m_full.predict(Xp)

    r2_full = r2_score(Y, yhat_full)
    ar2_full = r2_adjusted(r2_full, nobs=n, k=p)

    r2_full_list.append(r2_full)
    ar2_full_list.append(ar2_full)
```

```python
# ============================================================
# (2 puntos) Estimación con división train/test 75%/25% para R^2 out-of-sample
# ============================================================
for p in p_list:
    Xp = make_poly_features(X_scaled, p)
    Xtr, Xte, ytr, yte = train_test_split(Xp, Y, test_size=0.25, random_state=2025)

    m_split = LinearRegression(fit_intercept=False)
    m_split.fit(Xtr, ytr)
    yhat_test = m_split.predict(Xte)

    r2_oos = r2_score(yte, yhat_test)
    r2_oos_list.append(r2_oos)
```

### (5) Gráficos


```python
# ============================================================
# (3 puntos) Gráficos: R^2, R^2 ajustado, R^2 out-of-sample
# ============================================================

# Gráfico 1: R^2 en muestra completa
plt.figure()
plt.plot(p_list, r2_full_list, marker='o')
plt.xscale('log')
plt.xlabel('Número de características (p)')
plt.ylabel('R² en muestra')
plt.title('R² vs número de características (muestra completa)')
plt.grid(True)
plt.show()

# Gráfico 2: R^2 ajustado en muestra completa
plt.figure()
plt.plot(p_list, ar2_full_list, marker='o')
plt.xscale('log')
plt.xlabel('Número de características (p)')
plt.ylabel('R² ajustado en muestra')
plt.title('R² ajustado vs número de características (muestra completa)')
plt.grid(True)
plt.show()

# Gráfico 3: R^2 out-of-sample
plt.figure()
plt.plot(p_list, r2_oos_list, marker='o')
plt.xscale('log')
plt.xlabel('Número de características (p)')
plt.ylabel('R² fuera de muestra')
plt.title('R² fuera de muestra vs número de características')
plt.grid(True)
plt.show()

# Impresión rápida de resultados por si se requiere
for p, r2f, ar2f, r2o in zip(p_list, r2_full_list, ar2_full_list, r2_oos_list):
    print(f"p={str(p):>4} | R²={r2f: .4f} | R²_aj={ar2f: .4f} | R²_OOS={r2o: .4f}")
```

```python
# ============================================================
# Versión LASSO paralela al OLS del enunciado
# ============================================================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

rng = np.random.default_rng(123)

# Parámetros
n = 1000
sigma = 1.0
p_list = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]

# DGP: intercepto = 0
X = rng.standard_normal(n)
eps = sigma * rng.standard_normal(n)
Y = 1.0 * X + eps  # Y = X + ruido

# Escalar X a [-1, 1] para estabilidad en potencias
X_scaled = X / np.max(np.abs(X))

def make_poly_features(x_unit, p: int):
    powers = np.arange(1, p + 1, dtype=np.int32)
    return np.power(x_unit[:, None], powers, dtype=np.float64)  # sin intercepto

def r2_adjusted_k(r2: float, nobs: int, k_eff: int):
    denom = (nobs - k_eff - 1)
    if denom <= 0:
        return np.nan
    return 1.0 - (1.0 - r2) * (nobs - 1) / denom

# Contenedores
r2_full_lasso, ar2_full_lasso, r2_oos_lasso, k_eff_full = [], [], [], []

# ============================================================
# LASSO en muestra completa (para R² y R² ajustado con k_eff)
# ============================================================
for p in p_list:
    Xp = make_poly_features(X_scaled, p)

    # Estandarización en toda la muestra SOLO para este bloque "full"
    sc = StandardScaler(with_mean=True, with_std=True)
    Xp_std = sc.fit_transform(Xp)

    # LASSO con CV para elegir alpha
    lcv_full = LassoCV(cv=5, fit_intercept=False, random_state=123,
                       n_alphas=100, max_iter=5000)
    lcv_full.fit(Xp_std, Y)
    yhat_full = lcv_full.predict(Xp_std)

    r2f = r2_score(Y, yhat_full)
    k_eff = int(np.count_nonzero(lcv_full.coef_))  # grados de libertad efectivos
    ar2f = r2_adjusted_k(r2f, nobs=n, k_eff=k_eff)

    r2_full_lasso.append(r2f)
    ar2_full_lasso.append(ar2f)
    k_eff_full.append(k_eff)

# ============================================================
# LASSO con split 75/25 para R² fuera de muestra
# ============================================================
for p in p_list:
    Xp = make_poly_features(X_scaled, p)
    Xtr, Xte, ytr, yte = train_test_split(Xp, Y, test_size=0.25, random_state=2025)

    # Estandariza con estadísticos del train
    sc = StandardScaler(with_mean=True, with_std=True)
    Xtr_std = sc.fit_transform(Xtr)
    Xte_std = sc.transform(Xte)

    lcv = LassoCV(cv=5, fit_intercept=False, random_state=123,
                  n_alphas=100, max_iter=5000)
    lcv.fit(Xtr_std, ytr)
    yhat_test = lcv.predict(Xte_std)

    r2_oos_lasso.append(r2_score(yte, yhat_test))

# ============================================================
# Gráficos: R², R² ajustado (con k_eff), R² out-of-sample
# ============================================================
plt.figure()
plt.plot(p_list, r2_full_lasso, marker='o', label='LASSO (full)')
plt.xscale('log'); plt.xlabel('Número de características (p)')
plt.ylabel('R² en muestra'); plt.title('LASSO: R² vs p (muestra completa)')
plt.grid(True); plt.show()

plt.figure()
plt.plot(p_list, ar2_full_lasso, marker='o', label='LASSO (full)')
plt.xscale('log'); plt.xlabel('Número de características (p)')
plt.ylabel('R² ajustado (k_eff)'); plt.title('LASSO: R² ajustado vs p')
plt.grid(True); plt.show()

plt.figure()
plt.plot(p_list, r2_oos_lasso, marker='o', label='LASSO (test)')
plt.xscale('log'); plt.xlabel('Número de características (p)')
plt.ylabel('R² fuera de muestra'); plt.title('LASSO: R² test vs p')
plt.grid(True); plt.show()

# Resumen rápido
for p, r2f, ar2f, ke, r2o in zip(p_list, r2_full_lasso, ar2_full_lasso, k_eff_full, r2_oos_lasso):
    print(f"p={p:>4} | k_eff={ke:>4} | R²_full={r2f: .4f} | R²_aj(k_eff)={ar2f: .4f} | R²_OOS={r2o: .4f}")
```
